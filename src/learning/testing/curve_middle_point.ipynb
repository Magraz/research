```python
import numpy as np
import matplotlib.pyplot as plt

def find_middle_point(x, y):
    """
    Find the middle point of a curve defined by x, y points based on arc length.

    Parameters:
    x (array-like): x-coordinates of the curve
    y (array-like): y-coordinates of the curve

    Returns:
    (float, float): Middle point (x, y) of the curve
    """
    # Compute the distances between consecutive points
    distances = np.sqrt(np.diff(x)**2 + np.diff(y)**2)
    
    # Compute cumulative arc length
    cumulative_dist = np.insert(np.cumsum(distances), 0, 0)
    
    # Find the index where cumulative distance is closest to half of total arc length
    half_length = cumulative_dist[-1] / 2
    middle_index = np.argmin(np.abs(cumulative_dist - half_length))
    
    return x[middle_index], y[middle_index]

# Example Usage: Generate a random curve
t = np.linspace(0, 2 * np.pi, 50)
x = np.sin(t) + np.random.normal(0, 0.1, size=t.shape)  # Noisy sine wave
y = np.cos(t) + np.random.normal(0, 0.1, size=t.shape)

# Find the middle point
middle_x, middle_y = find_middle_point(x, y)

# Plot the curve and the middle point
plt.figure(figsize=(6,6))
plt.plot(x, y, 'b-', label="Curve")
plt.scatter(middle_x, middle_y, color='r', s=100, label="Middle Point")
plt.xlabel("X")
plt.ylabel("Y")
plt.legend()
plt.title("Finding the Middle Point of a Curve")
plt.grid()
plt.show()

print(f"Middle Point: ({middle_x:.4f}, {middle_y:.4f})")
```
